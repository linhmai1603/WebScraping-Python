```python
from urllib.request import urlopen
from bs4 import BeautifulSoup as bs
import pandas as pd
import requests
import sys
from PyQt5.QtCore import QUrl
from PyQt5.QtWidgets import QApplication
from PyQt5.QtWebEngineWidgets import QWebEnginePage
```

```python
job = "https://www.udel.edu/faculty-staff/human-resources/careers/faculty-jobs/"
```

```python
html = urlopen(job)
```

```python
page_content = bs(html, "html.parser")
```

```python
table = page_content.find_all('table')[2]
```

```python
tableTest = table.findAll('tr')
```

```python
jobs = []
```

```python
for tr in tableTest:
    td = tr.find_all('td')
    rows = [i.text.strip() for i in td]
    jobs.append(str(rows))
```

```python
jobs = [str(job).replace('\xa0',' ') for job in jobs]
jobs = [str(job).replace('\\xa0',' ') for job in jobs]
jobs = [str(job).replace("['",' ') for job in jobs]
jobs = [str(job).replace("']",' ') for job in jobs]
```

```python
links =[]
for aTag in table.findAll('a'):
    href = aTag.get('href')
    ##link = [i.text.strip() for i in href]
    ##rows = [i.text.strip() for i in href]
    links.append(str(href))
    
```

```python
jobs_list = pd.DataFrame({'Job Title': jobs, 'Link': links})
```

```python
class Client(QWebEnginePage):
    def __init__(self, url):
        self.app = QApplication(sys.argv)
        QWebEnginePage.__init__(self)
        self.html = ''
        self.loadFinished.connect(self._on_load_finished)
        self.load(QUrl(url))
        self.app.exec_()
    def _on_load_finished(self):
        self.html = self.toHtml(self.Callable)
        print('Load finished')

    def Callable(self, html_str):
        self.html = html_str
        self.app.quit()
url = Client("https://gems.fastmail.usf.edu:4440/psc/gemspro-tam/EMPLOYEE/HRMS/c/HRS_HRAM.HRS_APP_SCHJOB.GBL?Page=HRS_APP_SCHJOB&Action=U&FOCUS=Applicant&SiteId=1&PortalActualURL=https%3a%2f%2fgems.fastmail.usf.edu%3a4440%2fpsc%2fgemspro-tam%2fEMPLOYEE%2fHRMS%2fc%2fHRS_HRAM.HRS_APP_SCHJOB.GBL%3fPage%3dHRS_APP_SCHJOB%26Action%3dU%26FOCUS%3dApplicant%26SiteId%3d1&PortalRegistryName=EMPLOYEE&PortalServletURI=https%3a%2f%2fgems.fastmail.usf.edu%3a4440%2fpsp%2fgemspro-tam%2f&PortalURI=https%3a%2f%2fgems.fastmail.usf.edu%3a4440%2fpsc%2fgemspro-tam%2f&PortalHostNode=HRMS&NoCrumbs=yes&PortalKeyStruct=yes")
##client_response = Client(url)
##source = client_response.mainFrame().toHtml()
page_content = bs(url.html, "html.parser")
job_table = page_content.find("table", {"class" : "PSLEVEL1GRID"})
```

```python
trs = job_table.findAll('tr')
```

```python
title = []
job_ids = []
locations = []
departments =[]
job_families = []
dates = []
vp_areas = []
```

```python
for tr in trs:
    a = tr.find("a", {"class" : "PSHYPERLINKNOUL"})
    if(a is not None):
        title.append(str(a.text))
```

```python
len(title)
```

```python
len(trs)
```

```python
i = -1
for tr in trs:
    ##print(tr)
    job_id = tr.find("span", id =("JOBNUMBER$" +str(i)))    
    ##print(job_id)
    if(job_id is not None):
        job_ids.append(str(job_id.text))
    location = tr.find("span",id = "HRS_LOCATION_DESCR$" +str(i))     
    if(location is not None):
        locations.append(str(location.text))
    vp_area = tr.find("span", id = "HRS_CE_JO_EXT_I_U_VP_AREA$" +str(i))     
    if(vp_area is not None):
         vp_areas.append(str(vp_area.text))    
    department = tr.find("span",id = "U_PA_DEPT_VW_DESCR$164$$" +str(i))     
    if(department is not None):
        departments.append(str(department.text))
    job_family = tr.find("span", id = "JOB_FAMILY2$"+str(i))     
    if(job_family is not None):
        job_families.append(str(job_family.text))
    date = tr.find("span", id ="OPENED$" +str(i))     
    if(date is not None):
        dates.append(str(date.text))    
    i +=1  
```

```python
len(job_ids)
```

```python
jobs_list = pd.DataFrame({'Job Title': title, 'ID': job_ids, 'Location' : locations, 'VP area' : vp_areas, 'Department' : departments, 'Job Family' : job_families, 'Open Date' : dates})
```
